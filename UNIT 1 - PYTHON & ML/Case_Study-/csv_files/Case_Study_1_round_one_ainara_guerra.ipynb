#### 8.1. Delete the column education and the number of open complaints from the dataframe.

#### 8.2. Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].apply(lambda x: str(x))
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].apply(lambda x: x.replace("%", ""))
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].astype(float)
all_data['customer_lifetime_value'] = all_data['customer_lifetime_value'].replace(np.nan, 0)
all_data
```

```python
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].astype(str)
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].apply(lambda x: x.replace("NaN", "0"))
all_data["customer_lifetime_value"] = all_data["customer_lifetime_value"].astype(float)
all_data.head()
```

#### Plus. Other cleaning

```python
all_data.isna().sum()
```

```python
all_data['gender'].value_counts()
```

```python
all_data["gender"] = all_data["gender"].apply(lambda x: str(x))
all_data["gender"] = all_data["gender"].apply(lambda x: x.replace("Male", "M"))
all_data["gender"] = all_data["gender"].apply(lambda x: x.replace("female", "F"))
all_data["gender"] = all_data["gender"].apply(lambda x: x.replace("femal", "F"))
all_data["gender"] = all_data["gender"].apply(lambda x: x.replace("Femal", "F"))
all_data["gender"] = all_data["gender"].apply(lambda x: x.replace("nan", "Unknown"))
all_data['gender'].value_counts()
```

```python
all_data["customer"] = all_data["customer"].fillna("Unknown")
all_data["st"] = all_data["st"].fillna("Unknown")
all_data["income"] = all_data["income"].fillna("Unknown")
all_data["monthly_premium_auto"] = all_data["monthly_premium_auto"].fillna("Unknown")
all_data["policy_type"] = all_data["policy_type"].fillna("Unknown")
all_data["vehicle_class"] = all_data["vehicle_class"].fillna("Unknown")
all_data["total_claim_amount"] = all_data["total_claim_amount"].fillna("Unknown")
all_data.isna().sum()
```

```python
all_data["st"].value_counts()
```

```python
all_data["st"] = all_data["st"].apply(lambda x: str(x))
all_data["st"] = all_data["st"].apply(lambda x: x.replace("AZ", "Arizona"))
all_data["st"] = all_data["st"].apply(lambda x: x.replace("WA", "Washington"))
all_data["st"] = all_data["st"].apply(lambda x: x.replace("Cali ", "California"))
all_data["st"] = all_data["st"].apply(lambda x: x.replace("Californiafornia", "California"))
all_data["st"].value_counts()
```
