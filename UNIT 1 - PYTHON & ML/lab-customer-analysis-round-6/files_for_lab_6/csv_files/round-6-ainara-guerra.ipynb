# X-Y SPLIT

```python
y = df['total_claim_amount']
X = df.drop(['total_claim_amount'], axis=1)
x_num = df.select_dtypes(include = np.number)
```

# ONE HOT/ LABEL ENCODIDNG (categorical).


```python
# We are going to choose which one to encode taking into consideration that we are going to apply a linear regression: 
# we will encode response, coverage, employment_status, gender, policy_type, renew_offer_type
#we won't encode state, education, marital_status, policy, sales_channel, vehicle_class, vehicle_size
```

```python
response_enc = df['response']
response_enc_dumm = pd.get_dummies(response_enc, drop_first=False)
response_enc_dumm
```

```python
coverage_enc = df['coverage']
coverage_enc_dumm = pd.get_dummies(coverage_enc, drop_first=False)
coverage_enc_dumm
```

```python
employment_enc = df['employment_status']
employment_enc_dumm = pd.get_dummies(employment_enc, drop_first=False)
employment_enc_dumm
```

```python
gender_enc = df['gender']
gender_enc_dumm = pd.get_dummies(gender_enc, drop_first=False)
gender_enc_dumm
```

```python
policy_type_enc = df['policy_type']
policy_type_enc_dumm = pd.get_dummies(policy_type_enc, drop_first=False)
policy_type_enc_dumm
```

```python
renew_offer_type_enc = df['renew_offer_type']
renew_offer_type_enc_dumm = pd.get_dummies(renew_offer_type_enc, drop_first=False)
renew_offer_type_enc_dumm
```

# Concat DataFrames


```python
X_encoded = pd.concat ([x_num, response_enc_dumm, coverage_enc_dumm,employment_enc_dumm,gender_enc_dumm, policy_type_enc_dumm,renew_offer_type_enc_dumm], axis=1)
X_encoded = X_encoded.drop(['total_claim_amount'], axis=1)
X_encoded
```


# Linear Regression
#### Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)
```

#### Apply linear regression

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions  = model.predict(X_test)
predictions
```

#### Model validation

```python
r2 = r2_score(y_test, predictions)
RMSE = mean_squared_error(y_test, predictions, squared=False)
MSE = mean_squared_error(y_test, predictions)
MAE = mean_absolute_error(y_test, predictions)
print("r2 = ", r2)
print("RMSE = ", RMSE)
print("MSE = ", MSE)
print("MSE = ", MAE)
```
