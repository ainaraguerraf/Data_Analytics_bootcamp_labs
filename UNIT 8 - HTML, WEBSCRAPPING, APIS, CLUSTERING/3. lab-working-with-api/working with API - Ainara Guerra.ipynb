# <span style="color:rgb(255, 0, 255)"> Tripadvisor

Tripadvisor API helps to query realtime Hotels search, Flights prices, Restaurants, Attracting locations, etc to create a travelling site.

Tripadvisor, Inc. is an American online travel company that operates a website and mobile app with user-generated content and a comparison shopping website. It also offers online hotel reservations and bookings for transportation, lodging, travel experiences, and restaurants.

```python
url = "https://tripadvisor16.p.rapidapi.com/api/v1/flights/searchAirport"

querystring = {"query":"Madrid"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "tripadvisor16.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

# <span style="color:rgb(255, 0, 255)"> AI Trip Planner

The API generates personalized trip plans based on the user’s specified number of days and location. Using AI, the system recommends popular tourist attractions, local experiences, and accommodations based on user preferences and past travel behavior. The generated trip plans can be customized and adjusted to suit the user’s needs, making it a valuable tool for anyone planning a trip.

This is an API that generates personalized trip plans using artificial intelligence. It takes as input the number of days and the location of the trip, and outputs a detailed itinerary with recommended tourist attractions, local experiences, and accommodations.

To use this API, you will need an API key, which you can obtain by subscribing to this API. Once you have an API key, you can make HTTP requests to our API endpoint with the necessary parameters.

```python
url = "https://ai-trip-planner.p.rapidapi.com/"

querystring = {"days":"3","destination":"London,UK"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "ai-trip-planner.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

# <span style="color:rgb(255, 0, 255)"> Travel Advisor

This API helps to query realtime Flights prices, Hotels booking, Restaurants, Attracting locations, etc… to create a travelling site like tripadvisor.com

This API reproduce PUBLIC data and features of TripAdvisor - tripadvisor.com

```python
url = "https://travel-advisor.p.rapidapi.com/locations/v2/auto-complete"

querystring = {"query":"eiffel tower","lang":"en_US","units":"km"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "travel-advisor.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
url = "https://travel-advisor.p.rapidapi.com/attractions/v2/get-details"

querystring = {"currency":"USD","units":"km","lang":"en_US"}

payload = {
	"contentId": "1451754",
	"startDate": "2022-06-30",
	"endDate": "2022-07-01",
	"pax": [
		{
			"ageBand": "ADULT",
			"count": 2
		}
	]
}
headers = {
	"content-type": "application/json",
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "travel-advisor.p.rapidapi.com"
}

response = requests.post(url, json=payload, headers=headers, params=querystring)

print(response.json())
```
